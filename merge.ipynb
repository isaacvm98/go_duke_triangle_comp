```python
import pandas as pd
import numpy as np
import seaborn as sns
import zipfile
```

```python
spellings = pd.read_csv("data/MTeamSpellings.csv")
import re


def normalize_name(name):
    name = name.lower()
    name = re.sub(r"[^a-z0-9 ]", " ", name)  # remove punctuation, replace with space
    name = re.sub(r"\s+", " ", name).strip()  # collapse whitespace
    return name
```

```python
coaches = pd.read_csv("data/coaches.csv")
```

```python
coaches.info()
```

```python
spellings["norm"] = spellings["TeamNameSpelling"].apply(normalize_name)
coaches["norm"] = coaches["School"].apply(normalize_name)
```

```python
len(coaches)
```

```python
coaches = coaches.merge(spellings, on="norm").drop_duplicates()
```

```python
len(coaches)
```

```python
coaches = coaches.groupby(["Season", "norm", "Coach"]).last()
```

```python
coaches.to_csv("data/coaches_upd.csv")
```
